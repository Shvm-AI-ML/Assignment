```python
name = []
mrp = []
sp = []
Brand = []
url = []
weight = []
#country_of_origin = []
for items in total_data:
    name.append(items['p_desc'])
    mrp.append(items['mrp'])
    sp.append(items['sp'])
    Brand.append(items['p_brand'])
    url.append(items['absolute_url'])
    weight.append(items['w'])
    #country_of_origin.append(j['Country Of Origin'])
    #rating_count.append(j['rating_info'])
#printrp)
#print(p)
#rint(Brand)
#print(weight)
# Problem with the url add,www.bigbasket.com/
abs_url = []

for urls in url:
    urls = 'www.bigbasket.com' + urls
    abs_url.append(urls)
#print(len(abs_url))
```
